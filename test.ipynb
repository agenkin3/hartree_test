```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('dataset1.csv')
df1
```

```python
df2 = pd.read_csv('dataset2.csv')
df2
```

```python
df_merged = pd.merge(df1, df2, on=['counter_party'],suffixes=('_left', '_right'))
df_merged
```

```python
new_df = df_merged[['legal_entity', 'counter_party', 'tier', 'rating','status', 'value']]

max_rating_by_cp = new_df.groupby('counter_party')['rating'].max()
new_df['max(rating by counterparty)'] = new_df['counter_party'].map(max_rating_by_cp)
arap_sum = new_df.loc[new_df['status'] == 'ARAP', 'value'].sum()
accr_sum = new_df.loc[new_df['status'] == 'ACCR', 'value'].sum()
new_df['sum(value where status=ARAP)'] = new_df['status'].apply(lambda x: arap_sum if x == 'ARAP' else 0)
new_df['sum(value where status=ACCR)'] = new_df['status'].apply(lambda x: accr_sum if x == 'ACCR' else 0)
new_df.to_csv('new_record.csv')
```
